8.Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints
Greetings! if anything else is stored in spam.

```python
def check_value(spam):
    if spam == 1:
        print("Hello\n")
        #return "Hello"
    elif spam == 2:
        print("Howdy\n")
        #return "Howdy"
    else:
        print("Greetings!\n")
        #return "Greetings!"
        
# let spam = 1
spam = 1
print(f"spam = {spam}")
check_value(spam)

# let spam = 2
spam = 2
print(f"spam = {spam}")
check_value(spam)

# let spam = some data type except 1 and 2 numericals
spam = "some data type except 1 and 2 "
print(f"spam = {spam}")
check_value(spam)
```

12.Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.
